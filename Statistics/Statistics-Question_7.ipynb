## Answer-7 [Statistics] :-

```
The equations of the two lines of regression obtained in the correlation analysis are:

Line of regression of Y on X:  2𝑋 + 3 − 8 = 0

Line of regression of X on Y:  2𝑌 + 𝑋 − 5 = 0

Given: Variance of 𝑋 = 4

Ans-a) 

Variance of Y:

To find the variance of Y, we need to determine the slope of the line of regression of X on Y. From the 
equation 2𝑌 + 𝑋 − 5 = 0, we can rewrite it in terms of Y:


2𝑌 = −𝑋 + 5
𝑌 = −(1/2)𝑋 + 5/2


The slope of the line of regression of X on Y is −(1/2). Since the coefficient of determination (r^2) is 
the square of the correlation coefficient, which is the slope in this case, we can determine the variance 
of Y as:


Variance of Y = Variance of X × (slope of X on Y)^2
              = 4 × (−1/2)^2
              = 4 × 1/4
              = 1
Therefore, the variance of Y is 1.


Ans-b)

Coefficient of determination of X and Y:

The coefficient of determination (r^2) represents the proportion of the total variation in one variable (Y) 
that can be explained by the regression on the other variable (X).

Coefficient of determination of X and Y = (Variance of Y) / (Variance of X)
                                         = 1 / 4
                                         = 1/4

Therefore, the coefficient of determination of X and Y is 1/4.


Ans-c) Standard error of estimate of X on Y and of Y on X:

The standard error of estimate represents the average distance between the observed values and the predicted 
values from the regression line.

Standard error of estimate of X on Y = √(1 - r^2) × Standard deviation of X
                                     
                                     = √(1 - 1/4) × √4
                                     = √(3/4) × 2
                                     = √3
                                     

Standard error of estimate of Y on X = √(1 - r^2) × Standard deviation of Y
                                     = √(1 - 1/4) × √1
                                     = √(3/4)
                                     = √3/2


Therefore, the standard error of estimate of X on Y is √3, and the standard error of estimate of Y on X 
is √3/2.

```
